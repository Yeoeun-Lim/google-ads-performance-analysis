```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.metrics import calculate_kpis

plt.rcParams['figure.figsize'] = (10,6)
```

```python
df = pd.read_csv('../data/cleaned_ads_data.csv')

df.head()
```

```python
df.info()
df.describe()
```

```python
df = calculate_kpis(df)

df[['CTR', 'Conversion_Rate', 'ROI']].describe()
```

## Distribution 분석

```python
# Clicks / Impressions

sns.histplot(df['Clicks'], bins=50)
plt.title('Clicks Distribution')
plt.show()

sns.histplot(df['Impressions'], bins=50)
plt.title('Impressions Distribution')
plt.show()
```

```python
# Conversion

sns.histplot(df['Conversions'], bins=50)
plt.title('Conversions Distribution')
plt.show()
```

```python
sns.histplot(df['CTR'], bins=50)
plt.title('CTR Distribution')
plt.show()

sns.histplot(df['Conversion_Rate'], bins=50)
plt.title('Conversion Rate Distribution')
plt.show()
```

## Relationship 분석

```python
# Cost vs Conversions

sns.scatterplot(x='Cost', y='Conversions', data=df)
plt.title('Cost vs Conversions')
plt.show()
```

```python
# Clicks vs Conversions

sns.scatterplot(x='Clicks', y='Conversions', data=df)
plt.title('Clicks vs Conversions')
plt.show()
```

```python
# CTR vs Conversion Rate

sns.scatterplot(x='CTR', y='Conversion_Rate', data=df)
plt.title('CTR vs Conversion Rate')
plt.show()
```
